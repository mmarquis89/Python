# Calculate scaled sum of expression values of DE genes and plot

```python
def calc_expression_score(df, names, genesA, genesB):
    # Calculates expression score defined as the scaled sum of expression values of genesA minus the same for genesB
    Z = copy.deepcopy(df[names].loc[genesA])
    scoresA = Z.div(Z.max(axis=1), axis=0).mean(axis=0)
    Z = copy.deepcopy(df[names].loc[genesB])
    scoresB = Z.div(Z.max(axis=1), axis=0).mean(axis=0)
    return scoresB - scoresA
```

```python
# Using only 24, 36, and 48h APF
names_adPN = list(names_adPN_ClassA) + list(names_adPN_ClassB)
scores_adPN_ClassA_ClassB = calc_expression_score(X, names_adPN, genes_DE_ClassA_ClassB_ClassAHigh, genes_DE_ClassA_ClassB_ClassBHigh)
scores_adPN_24hAPF_adult = calc_expression_score(X, names_adPN, genes_DE_24hAPF_adult_24hAPFHigh, genes_DE_24hAPF_adult_adultHigh)
```

```python
# Using all times
scores_adPN_all_ClassA_ClassB = calc_expression_score(X, names_adPN_all, genes_DE_ClassA_ClassB_ClassAHigh, genes_DE_ClassA_ClassB_ClassBHigh)
scores_adPN_all_24hAPF_adult = calc_expression_score(X, names_adPN_all, genes_DE_24hAPF_adult_24hAPFHigh, genes_DE_24hAPF_adult_adultHigh)
```

# Correlation of ICIM genes

```python
def calc_pairwise_corr(X, Y, ignore_self=False, genes=None):
    # Calculate correlation between each column of X and each column of Y
    corrs = []
    if genes is not None:
        X_clean = X.loc[genes]
        Y_clean = Y.loc[genes]
    else:
        X_clean = X
        Y_clean = Y
    for name1 in X_clean:
        for name2 in Y_clean:
            if ignore_self and name1 == name2: continue
            corr, pvalue = scipy.stats.pearsonr(X_clean[name1], Y_clean[name2])
            corrs.append(corr)
    return corrs
```

```python
names_adPN_all_24hAPF = df_libs.loc[names_adPN_all].loc[df_libs.loc[names_adPN_all]["label"] == "Mz19-GFP,24hAPF"].index
names_adPN_all_36hAPF = df_libs.loc[names_adPN_all].loc[df_libs.loc[names_adPN_all]["label"] == "Mz19-GFP,36hAPF"].index
names_adPN_all_48hAPF = df_libs.loc[names_adPN_all].loc[df_libs.loc[names_adPN_all]["label"] == "Mz19-GFP,48hAPF"].index
names_adPN_all_72hAPF = df_libs.loc[names_adPN_all].loc[df_libs.loc[names_adPN_all]["label"] == "Mz19-GFP,72hAPF"].index
names_adPN_all_adult = df_libs.loc[names_adPN_all].loc[df_libs.loc[names_adPN_all]["label"] == "Mz19-GFP,adult"].index
```

```python
corrs_adPN_all_24hAPF_ICIM = calc_pairwise_corr(X[names_adPN_all_24hAPF], X[names_adPN_all_24hAPF], genes=genes_Mz19Development_ICIM, ignore_self=True)
corrs_adPN_all_36hAPF_ICIM = calc_pairwise_corr(X[names_adPN_all_36hAPF], X[names_adPN_all_36hAPF], genes=genes_Mz19Development_ICIM, ignore_self=True)
corrs_adPN_all_48hAPF_ICIM = calc_pairwise_corr(X[names_adPN_all_48hAPF], X[names_adPN_all_48hAPF], genes=genes_Mz19Development_ICIM, ignore_self=True)
corrs_adPN_all_72hAPF_ICIM = calc_pairwise_corr(X[names_adPN_all_72hAPF], X[names_adPN_all_72hAPF], genes=genes_Mz19Development_ICIM, ignore_self=True)
corrs_adPN_all_adult_ICIM = calc_pairwise_corr(X[names_adPN_all_adult], X[names_adPN_all_adult], genes=genes_Mz19Development_ICIM, ignore_self=True)
```

```python
fig, ax = plt.subplots(1, 1, figsize=(7,5.5))
palette = ["#e6f0ff", "#99c2ff", "#4d94ff", "#0066ff", "#003C96"] # colors for points
sns.violinplot(data=[corrs_adPN_all_24hAPF_ICIM,
                     corrs_adPN_all_36hAPF_ICIM,
                     corrs_adPN_all_48hAPF_ICIM,
                     corrs_adPN_all_72hAPF_ICIM,
                     corrs_adPN_all_adult_ICIM],
               cut=0,
               palette=palette,
               ax=ax)
ax.set_ylim(top=0.8)
ax.set_ylabel("Similarity (Pearson correlation of ICIM genes)")
ax.set_xticklabels(["24 hr APF", "36 hr APF", "48 hr APF", "72 hr APF", "adult"])
plt.tight_layout()
```
